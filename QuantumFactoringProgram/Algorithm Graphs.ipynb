```python
# number of bits L
# number N
L = np.arange(100)

x_vals = np.delete(np.delete(L, 0), 0)
y_vals = [x for x in x_vals]

fig, ax = plt.subplots()
ax.plot(x_vals, y_vals, linewidth=3, linestyle='solid')


#O(L2log L log log L)
y_vals = [sqrt(x) for x in x_vals]
ax.plot(x_vals, y_vals, linewidth=3, linestyle='solid')

ax.set(xlabel='Number of Items in List (N)', ylabel='Steps')

plt.annotate(text='Grover\'s', xy=(60, 20))
plt.annotate(text='General\n Search', xy=(20, 40))
```
